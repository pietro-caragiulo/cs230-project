```python
import os 
from os.path import dirname
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')

from DataParser import DataParser
from CsvParser  import CsvParser
```

## Datasets location

Manually type where the signal and background are located

```python
data_dir        = os.path.join(os.getcwd(), '../datasets')

signal_file_orig     = 'ap_100MeV_L1L1_loose.csv'
background_file_orig = 'tritrig-wab-beam_100MeV_L1L1_loose.csv'

signal_file_02mm     = './02mm/signal_02mm.csv'
background_file_02mm = './02mm/background_02mm.csv'

signal_file_08mm     = './08mm/signal_08mm.csv'
background_file_08mm = './08mm/background_08mm.csv'

signal_file_2mm     = './2mm/signal_2mm.csv'
background_file_2mm = './2mm/background_2mm.csv'

```

## Background, Signal and Data object declaration

CsvParser objects attributes map the CSV structure
DataParser objects combine the signal and background CsvParser objects, add a label and shuffle the data

```python
background_orig = CsvParser(os.path.join(data_dir, background_file_orig))
signal_orig     = CsvParser(os.path.join(data_dir, signal_file_orig))

background_02mm = CsvParser(os.path.join(data_dir, background_file_02mm))
signal_02mm     = CsvParser(os.path.join(data_dir, signal_file_02mm))

background_08mm = CsvParser(os.path.join(data_dir, background_file_08mm))
signal_08mm     = CsvParser(os.path.join(data_dir, signal_file_08mm))

background_2mm = CsvParser(os.path.join(data_dir, background_file_2mm))
signal_2mm     = CsvParser(os.path.join(data_dir, signal_file_2mm))

myData_orig     = DataParser(signal=signal_orig, background=background_orig)
myData_02mm     = DataParser(signal=signal_orig, background=background_02mm)
myData_08mm     = DataParser(signal=signal_orig, background=background_08mm)
myData_2mm      = DataParser(signal=signal_orig, background=background_2mm)
```

## Plot Histograms of Signal and Background

Visually inspect the plots for now to check if the plot label matches between signal and background

```python
# Set-up picture
fig_row = 6
fig_col = 6
fig, ax = plt.subplots(fig_row, fig_col, figsize=(15,15))
fig.tight_layout()

signal = signal_orig
background = background_orig
# Plot background
for i, feature in enumerate(background.CSV_HEADER):    
    thisAx = ax[i % fig_col, i // fig_row]
    _ = thisAx.hist(background.RAW_DF[feature], bins=50, label=feature)
    thisAx.legend()
    thisAx.set_xticks([]) 
    thisAx.set_yticks([])

# Plot signal    
for i, feature in enumerate(signal.CSV_HEADER):    
    thisAx = ax[i % fig_col, i // fig_row]
    _ = thisAx.hist(signal.RAW_DF[feature], bins=50, label=feature)
    thisAx.legend()
    thisAx.set_xticks([]) 
    thisAx.set_yticks([])  
# Remove extra axies
for j in range(i+1, fig_row*fig_col):
    fig.delaxes(ax[j % fig_col, j // fig_row])

plt.show()
```
